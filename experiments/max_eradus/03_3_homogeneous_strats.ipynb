```python
import itertools

import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
from IPython.display import display

from ant.centralised import P4
from ant.decentralised.direct import GreedyAgent, ProportionalAgent
from ant.decentralised.special_strategies import MixedAgent
from ant.market import BaseAgent, Market
```

```python
market_size = 20
simulation_length = 1000
num_repeats = 30

graph_types = ["complete", "random", "grid", "scale-free", "small-world"]

strategy_mixes = {
    "Purely Greedy": {"strategies": [GreedyAgent], "weights": [1.0]},
    "Purely Proportional": {"strategies": [ProportionalAgent], "weights": [1.0]},
    "50/50 Greedy/Proportional": {"strategies": [GreedyAgent, ProportionalAgent], "weights": [0.5, 0.5]}
}

all_results = {
    mix_name: {gtype: [] for gtype in graph_types} for mix_name in strategy_mixes.keys()
}
```

```python
for repeat in range(num_repeats):
    print(f"Running repeat {repeat + 1}/{num_repeats}...")
    current_seed = 657849724 + repeat

    rnd_seed = current_seed
    random_graph = nx.gnp_random_graph(market_size, 0.2, seed=rnd_seed)
    while not nx.is_connected(random_graph):
        rnd_seed += 1000
        random_graph = nx.gnp_random_graph(market_size, 0.2, seed=rnd_seed)

    graphs = {
        "complete": nx.complete_graph(market_size),
        "random": random_graph,
        "grid": nx.convert_node_labels_to_integers(nx.grid_2d_graph(int(market_size / 4), 4)),
        "scale-free": nx.barabasi_albert_graph(market_size, 1, seed=current_seed),
        "small-world": nx.connected_watts_strogatz_graph(market_size, int(market_size / 2), 0.05, seed=current_seed),
    }

    for mix_name, mix_info in strategy_mixes.items():
        for graph_type, graph in graphs.items():
            mixed_agents = [
                MixedAgent(
                    id=i,
                    seed=current_seed + i,
                    strategies=mix_info["strategies"],
                    weights=mix_info["weights"],
                )
                for i in range(market_size)
            ]

            _market = Market(
                market_size,
                graph=graph,
                agents=mixed_agents,
                seed=current_seed,
                agent_type=BaseAgent,
            )

            for agent in _market.agents:
                agent.market = _market
                agent.received = np.zeros(market_size, dtype=float)
                agent.received_history = np.zeros((10000, market_size), dtype=float)

            optimal_aloc, utility_vec = P4(_market)
            _market.set_market_equilibrium(optimal_aloc, utility_vec)

            simulated_state = _market.simulate(simulation_length)
            all_results[mix_name][graph_type].append(simulated_state)
```

```python
training_period = 15
time_steps = np.arange(training_period, simulation_length)
colors = plt.cm.tab10.colors

plt.figure(figsize=(10, 6))

for idx, mix_name in enumerate(all_results.keys()):
    all_topologies_average = []

    for graph_type in graph_types:
        states_array = np.array(all_results[mix_name][graph_type])
        if states_array.size == 0:
            continue
        avg_state = np.mean(states_array, axis=0)
        all_topologies_average.append(avg_state)

    if len(all_topologies_average) > 0:
        grand_average = np.mean(all_topologies_average, axis=0)

        plt.plot(
            time_steps,
            grand_average[training_period:],
            label=mix_name,
            color=colors[idx % len(colors)],
            linewidth=2,
            alpha=0.9,
        )

plt.xlabel("Time Steps")
plt.ylabel("Loss")
plt.title(f"Averaged Over {num_repeats} Repeats & All Topologies")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
for graph_type in graph_types:
    plt.figure(figsize=(10, 6))

    for idx, mix_name in enumerate(all_results.keys()):
        states_array = np.array(all_results[mix_name][graph_type])
        if states_array.size == 0:
            continue
        avg_state = np.mean(states_array, axis=0)

        plt.plot(
            time_steps,
            avg_state[training_period:],
            label=mix_name,
            color=colors[idx % len(colors)],
            linewidth=2,
            alpha=0.8,
        )

    plt.xlabel("Time Steps")
    plt.ylabel("Loss")
    plt.title(f"{graph_type.capitalize()} Topology (Avg of {num_repeats} repeats)")
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
summary_data = []

for mix_name in all_results.keys():
    all_topologies_average = []

    for graph_type in graph_types:
        states_array = np.array(all_results[mix_name][graph_type])
        if states_array.size == 0:
            continue
        avg_state = np.mean(states_array, axis=0)
        all_topologies_average.append(avg_state)

    if len(all_topologies_average) == 0:
        final_avg_loss = float("nan")
    else:
        grand_average = np.mean(all_topologies_average, axis=0)
        final_avg_loss = np.mean(grand_average[training_period:])

    summary_data.append({
        "Strategy Mix": mix_name,
        "Average Loss": final_avg_loss,
    })

summary_df = pd.DataFrame(summary_data).sort_values("Average Loss")
print("Average Loss Over All Topologies")
display(summary_df)
```

```python
topology_data = []

for mix_name in all_results.keys():
    for graph_type in graph_types:
        states_array = np.array(all_results[mix_name][graph_type])
        if states_array.size == 0:
            final_loss = float("nan")
        else:
            avg_state = np.mean(states_array, axis=0)
            final_loss = np.mean(avg_state[training_period:])

        topology_data.append({
            "Strategy Mix": mix_name,
            "Topology": graph_type.capitalize(),
            "Average Loss": final_loss,
        })

topology_df = pd.DataFrame(topology_data)
pivot_df = topology_df.pivot(index="Strategy Mix", columns="Topology", values="Average Loss")

print("Average Loss for Each Topology")
display(pivot_df)
```
