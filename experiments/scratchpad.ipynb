```python
import numpy as np
```

```python
I_2 = np.array([
    [1, 0],
    [0, 1]
])
Z = np.array([
    [1, 0],
    [0, -1]
])
X = np.array([
    [0, 1],
    [1, 0]
])
```

```python
def is_unitary(M):
    return np.allclose(np.eye(len(M)), M.dot(M.T.conj()))
```

```python
np.kron(I_2, I_2)
```

```python
A_I = np.kron(I_2, np.kron(I_2, I_2))
A_R = np.kron(Z, np.kron(X, X))

A = 1/2 * (A_I - A_R)
print(A_I)
print(A_R)
print(A)
```

```python
print(A - A.T)
print(A @ A.T)
print(A @ A.T - A)
```

```python
print(is_unitary(A), is_unitary(A_I), is_unitary(A_R))
```
